```python
import pandas as pd #Customer Lifetime Value
```

```python
Df1 = pd.read_csv('files_for_lab/csv_files/file1.csv')
Df1['Customer Lifetime Value'] = Df1['Customer Lifetime Value'].str.rstrip('%').astype(float) / 100

Df1
```

```python
Df2 = pd.read_csv('files_for_lab/csv_files/file2.csv')
Df2['Customer Lifetime Value'] = Df2['Customer Lifetime Value'].str.rstrip('%').astype(float) / 100
Df2
```

```python
Df3 = pd.read_csv('files_for_lab/csv_files/file3.csv')
Df3
```

```python
Df1.columns = Df1.columns.str.capitalize()
Df1 = Df1.rename(columns={'St':'State'})

Df2.columns = Df2.columns.str.capitalize()
Df2 = Df2.rename(columns={'St':'State'})
```

```python
cols = []

for i in range(len(Df3.columns)):
    cols.append(Df3.columns[i].capitalize()) #mirar esto
    
Df3.columns = cols

Df3.head()
```

```python
total_claim = Df2['Total claim amount']

Df2 = Df2.drop(columns = ['Total claim amount']) #reorganizamos dataframe 2 
Df2.insert(loc = 10, column = 'Total claim amount', value = total_claim) #borrando e insertando misma columna en sitio elegido

Df3 = Df3.reindex(columns=['Customer', 'State', 'Gender', 'Education', 'Customer lifetime value', 'Income',
                           'Monthly premium auto', 'Number of open complaints', 'Policy type', 'Vehicle class',
                           'Total claim amount']) #reorganizamos columnas dataframe 3

                           
```

```python
data = pd.concat([Df1, Df2, Df3])

data
```
